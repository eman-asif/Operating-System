### **Task 6: CPU Affinity**
CPU affinity binds a process to specific CPU cores. Two types are **hard affinity** (strict binding) and **soft affinity** (preferred but not mandatory). Use `taskset` or `/proc/[PID]/status` to explore this.

---

### **Task 7: Schedtool Parameters**
Use:
```bash
schedtool -n -p <PID>
schedtool -l
```
to list scheduling policies.

### **Task 8: Change Scheduling Parameters**
Commands for experiments:
- Policy: `schedtool -B -p <PID>` (Batch), `-I` (Idle), `-N` (Normal).
- Nice Value: `renice`.
- Static Priority: `schedtool`.
- CPU Affinity: `taskset`.

### **Task 9: `/proc` Filesystem**
Answering some key questions:
- **Why a window to the kernel?**: It provides dynamic runtime information about processes and system status.
- **File sizes**: They appear as 0 because they’re dynamically generated.
- **File descriptions**:
  - `cmdline`: Command-line arguments of the process.
  - `environ`: Environment variables.
  - `limits`: Resource limits.
  - `status`: Process metadata.

For the rest, explore files like `/proc/cpuinfo` for CPU info, `/proc/uptime`, and `/proc/sys/kernel`.
